```python
# 3차원(입체) 그래프를 그리기 위한 도구
from mpl_toolkits.mplot3d import Axes3D
# PCA: 여러 개의 특성을 중요한 것만 뽑아주는 도구
from sklearn.decomposition import PCA


# 그림(캔버스) 생성, 크기 설정
fig=plt.figure(figsize=(8,6))
# 3D(입체) 그래프를 그릴 축 생성
ax=fig.add_subplot(111,projection='3d')
# 3D 그래프를 바라보는 각도 설정
ax.view_init(elev=-150,azim=110)
# PCA를 이용해 차원을 3개로 줄이기
# target(정답)은 제외하고 feature만 사용
X_reduced=PCA(n_components=3).fit_transform(df_iris.drop('target',axis=1))
# PCA로 줄인 데이터로 3D 산점도 그리기
ax.scatter(X_reduced[:,0],  # 첫 번째 PCA 축
           X_reduced[:,1], # 두 번째 PCA 축
           X_reduced[:,2],# 세 번째 PCA 축
           c=df_iris['target'], # 색깔은 꽃 종류
           cmap=plt.cm.Set1,  # 색상 테마
           edgecolors='k',# 점 테두리 검정색
           s=40)   # 점 크기

# 그래프 제목
ax.set_title('Iris 3D')
# x축 이름
ax.set_xlabel('x')
```

```python
# 데이터를 훈련용(train)과 검증용(valid)으로 나누는 함수 불러오기
from sklearn.model_selection import train_test_split
# df_iris 데이터를 랜덤으로 섞어서
# 75%는 학습용, 25%는 검증용으로 나눔
x_train,x_valid,y_train,y_valid=train_test_split(df_iris.drop('target',axis=1), # 입력 데이터 (문제)
                                                 df_iris['target']) # 정답 데이터
```

```python
# 학습용 데이터의 크기 확인
# (데이터 개수, feature 개수)
x_train.shape,y_train.shape
```

```python
# 검증용 데이터의 크기 확인
x_valid.shape,y_valid.shape
```

```python
# seaborn 라이브러리를 사용해
# 각 클래스(정답)의 개수를 막대그래프로 표시

sns.countplot(x=y_train)# 학습용 정답 데이터 (꽃 종류 0,1,2)

# 그래프를 화면에 보여주기
plt.show()
```

```python
# 데이터를 학습용(train)과 검증용(valid)으로 나누기
# stratify를 사용해 클래스 비율을 그대로 유지
x_train,x_valid,y_train,y_=train_test_split(df_iris.drop('target',axis=1), # 입력 데이터 (문제)
                                            df_iris['target'],# 정답 데이터
                                            stratify=df_iris['target']) # 클래스 비율 유지
```

```python
# 학습용 정답 데이터의 분포 확인
# (클래스가 골고루 있는지 체크)
sns.countplot(x=y_train)
plt.show()
```

```python
# 학습용 데이터 크기 확인
x_train.shape,y_train.shape
```

```python
# 검증용 데이터 크기 확인
x_valid.shape,y_valid.shape
```

```python
from sklearn.linear_model import LogisticRegression

model=LogisticRegression()
```

```python
# 모델 학습 (공부시키기)
model.fit(x_train,y_train)
```

```python
# 검증용 데이터로 예측 수행
prediction=model.predict(x_valid)
```

```python
# 예측 결과 앞 5개만 확인
prediction[:5]
```

```python
# 정확도 계산
# (맞았으면 1, 틀리면 0 → 평균)
(prediction == y_valid).mean()
```

# 서포트벡터 머신(SVC)

- 새로운 데이터가 어느 카테고리에 속할지 판단하는 비확률적 이진 선형 분류 모델을 만듦

```python
# SVM(Support Vector Machine) 분류기 불러오기
from sklearn.svm import SVC

# SVM 모델 생성
# random_state를 고정해 실행할 때마다 같은 결과가 나오게 함
svc=SVC(random_state=0,)

# 학습용 데이터로 모델 학습
# 꽃 특징과 정답을 보고 최적의 경계선을 찾음
svc.fit(x_train,y_train)

# 검증용 데이터로 예측 수행
svc_pred=svc.predict(x_valid)
```

```python
# 예측 결과와 실제 정답을 비교해 정확도 계산
# 맞으면 1, 틀리면 0 → 평균 = 정확도
(svc_pred == y_valid).mean()
```

```python
# 예측 결과 중 앞 5개만 확인
svc_pred[:5]
```

```python
# 각 데이터가 결정 경계선에서 얼마나 떨어져 있는지 점수 확인
# 값이 클수록 해당 클래스라고 확신함
svc.decision_function(x_valid)[:5]
```

## Error(오차)

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# 데이터 로드
cancer = load_breast_cancer()

# 입력값과 정답 분리
data=cancer['data']
target=cancer['target']
feature_names=cancer['feature_names']
```

```python
# DataFrame 생성
df=pd.DataFrame(data=data,columns=feature_names)
df['target']=cancer['target']

df.head()
```

```python
# 양성(1) / 악성(0) 분리
pos=df.loc[df['target']==1]

neg=df.loc[df['target']==0]
```

```python
pos
```

```python
neg
```

```python
# 악성 데이터 5개만 사용 (불균형 데이터 만들기)
sample=pd.concat([pos,neg[:5]],sort=True)

sample
```

```python
# 학습 / 테스트 데이터 분리
x_train,x_test,y_train,y_test=train_test_split(sample.drop('target',axis=1),sample['target'],random_state=42)
```

```python
# 로지스틱 회귀 모델 생성 및 학습
model=LogisticRegression()
model.fit(x_train,y_train)

# 예측
pred=model.predict(x_test)
```

```python
# 모델 정확도
(pred == y_test).mean()
```

```python
# 모두 1로 예측하는 가짜 모델
my_prediction=np.ones(shape=y_test.shape)
```

```python
# 가짜 모델 정확도
(my_prediction == y_test).mean()
```

# 오차행렬

```python
# 오차행렬 계산 함수 불러오기
from sklearn.metrics import confusion_matrix

# 실제 정답(y_test)과 예측값(pred)을 비교해서 오차행렬 생성
confusion_matrix(y_test,pred)
```

```python
# 오차행렬을 히트맵으로 시각화
sns.heatmap(confusion_matrix(y_test,pred)# 오차행렬 값
            ,annot=True,# 각 칸에 숫자 표시
            cmap='Reds',)# 빨간색 계열 색상 사용

# x축 이름: 모델이 예측한 값
plt.xlabel('Predict')
# y축 이름: 실제 정답
plt.ylabel('Actual')
# 그래프 화면에 출력
plt.show()
```

```python
# 정밀도(Precision), 재현율(Recall) 계산 함수 불러오기
from sklearn.metrics import precision_score,recall_score

# Precision 계산
# → 모델이 1이라고 예측한 것 중, 진짜 1의 비율
precision_score(y_test,pred)
```

```python
# Recall 계산
# → 실제 1 중에서 모델이 맞게 1이라고 예측한 비율
recall_score(y_test,pred)
```
