## 교차검증

- K 폴드


```python
from sklearn.tree import DecisionTreeClassifier  # 결정트리 분류기
from sklearn.metrics import accuracy_score # 정확도 계산 함수
from sklearn.model_selection import KFold # KFold 교차검증 도구
import numpy as np # 숫자 계산용 라이브러리

# 붓꽃 데이터 불러오기
iris=load_iris()

# 꽃의 특징 데이터 (입력값)
features=iris.data

# 꽃의 종류 (정답)
label=iris.target

# 결정트리 모델 생성 (결과 고정을 위해 random_state 설정)
dt_clf=DecisionTreeClassifier(random_state=156)

# 데이터를 5등분하는 KFold 객체 생성
kfold=KFold(n_splits=5)
# 각 폴드의 정확도를 저장할 리스트
cv_accuracy=[]

print("붓꽃 데이터 세트 크기:",features.shape[0])
```

```python
# 교차검증 반복 횟수 카운트용 변수
n_iter=0


# KFold로 데이터 나누기 시작
for train_index,test_index in kfold.split(features):
    # 인덱스를 이용해 학습용/검증용 데이터 나누기
    X_train,X_test=features[train_index],features[test_index]
    y_train,y_test=label[train_index],label[test_index]
    # 결정트리 모델 학습
    dt_clf.fit(X_train,y_train)
    # 검증 데이터로 예측 수행
    pred=dt_clf.predict(X_test)
    # 반복 횟수 증가
    n_iter +=1
    # 정확도 계산 (소수점 4자리)
    accuracy=np.round(accuracy_score(y_test,pred),4)
    # 학습/검증 데이터 개수
    train_size=X_train.shape[0]
    test_size=X_test.shape[0]
    # 결과 출력
    print("\n{0} 교차검증 정확도 :{1},학습데이터크기:{2}, 검증 데이터 크기 {3}".format(n_iter,accuracy,train_size,test_size))
    # 검증에 사용된 데이터 인덱스 출력
    print("{0} 검증세트인덱스 :{1}".format(n_iter,test_index))
    # 정확도 저장
    cv_accuracy.append(accuracy)
# 전체 교차검증 평균 정확도
print("\n평균 검증 정확도:",np.mean(cv_accuracy))
```

## Stratified K폴드

```python
import pandas as pd
# 1. 붓꽃 데이터 불러오기
iris=load_iris()

# 2. 붓꽃 데이터를 DataFrame(표)로 만들기
iris_df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
# 3. 정답(label) 컬럼 추가
iris_df['label']=iris.target
# 4. 전체 데이터의 레이블 분포 확인
iris_df['label'].value_counts()
```

```python
# 5. KFold 설정 (3등분)
kfold=KFold(n_splits=3)
# 교차검증 반복 횟수
n_iter=0
for train_index,test_index in kfold.split(iris_df):
    n_iter +=1
    # 학습용 / 검증용 레이블 나누기
    label_train=iris_df['label'].iloc[train_index]
    label_test=iris_df['label'].iloc[test_index]
    print("## 교차검증 : {0}".format(n_iter))
    # 학습 데이터의 레이블 분포 출력
    print("학습레이블 데이터 분포: \n",label_train.value_counts())
    # 검증 데이터의 레이블 분포 출력
    print("검증 레이블 데이터 분포: \n",label_test.value_counts())
```

```python
from sklearn.model_selection import StratifiedKFold
# StratifiedKFold 불러오기
# → 라벨(label) 비율을 유지하면서 데이터를 나눠주는 교차검증 도구

# 1. StratifiedKFold 객체 생성 (3등분)
skf=StratifiedKFold(n_splits=3)

# 데이터를 3개로 나눠서
# 총 3번의 교차검증을 하겠다는 뜻

# 교차검증 반복 횟수를 세기 위한 변수
n_iter=0

# 2. 교차검증 시작
# split(데이터, 라벨)
# 두 번째 인자로 label을 넣어야
# 각 fold마다 라벨 비율이 동일하게 유지됨
for train_index,test_index in skf.split(iris_df,iris_df['label']):
    # 현재 몇 번째 교차검증인지 카운트 증가
    n_iter +=1

    # train_index에 해당하는 라벨만 뽑기 (학습용 라벨
    label_train=iris_df['label'].iloc[train_index]
    # test_index에 해당하는 라벨만 뽑기 (검증용 라벨)
    label_test=iris_df['label'].iloc[test_index]
    # 교차검증 번호 출력
    print("## 교차검증 :{0}".format(n_iter))
    # 학습 데이터에 포함된 각 라벨(클래스) 개수 출력
    print("학습 레이블 데이터 분포:\n",label_train.value_counts())
    # 검증 데이터에 포함된 각 라벨(클래스) 개수 출력
    print("검증 레이블 데이터 분포:\n",label_test.value_counts())
```

```python
# 결정트리 분류 모델 생성
# random_state를 고정하면 실행할 때마다 같은 결과가 나오게 됨
dt_clf=DecisionTreeClassifier(random_state=156)

# StratifiedKFold 객체 생성
# 데이터를 3개로 나누되, 각 클래스(label) 비율을 유지하면서 나눔
skfold=StratifiedKFold(n_splits=3)
# 현재 몇 번째 교차검증인지 세기 위한 변수
n_iter=0
# 각 교차검증에서 나온 정확도를 저장할 리스트
cv_accuracy=[]

# StratifiedKFold를 이용한 교차검증 시작
# features : 입력 데이터(X)
# label    : 정답 데이터(y) → 이걸 넣어야 클래스 비율이 유지됨
for train_index,test_index in skfold.split(features,label):
    # 학습용 데이터와 검증용 데이터로 나누기
    # train_index에 해당하는 데이터는 학습용
    # test_index에 해당하는 데이터는 검증용
    X_train,X_test=features[train_index],features[test_index]
    y_train,y_test=label[train_index],label[test_index]

    # =====================
    # 모델 학습 단계
    # =====================
    # 학습 데이터(X_train)와 정답(y_train)을 이용해 모델을 학습시킴
    dt_clf.fit(X_train,y_train)
    # =====================
    # 모델 예측 단계
    # =====================
    # 학습이 끝난 모델로 검증 데이터(X_test)를 예측
    pred=dt_clf.predict(X_test)
    
    # 교차검증 반복 횟수 증가
    n_iter +=1

    # 예측 결과(pred)와 실제 정답(y_test)을 비교해 정확도 계산
    # 소수점 4자리까지 반올림
    accuracy=np.round(accuracy_score(y_test,pred),4)
    # 학습 데이터와 검증 데이터의 개수 확인

    train_size=X_train.shape[0] # 학습 데이터 개수
    test_size=X_test.shape[0]  # 검증 데이터 개수
    # 현재 교차검증 결과 출력
    print("\n{0} 교차검증 정확도 :{1},학습데이터 크기: {2},검증데이터크기: {3}".format(n_iter,accuracy,train_size,test_size))

    print("{0} 검증세트인덱스 :{1}".format(n_iter,test_index))
    cv_accuracy.append(accuracy)

print("\n## 교차 검증별 정확도:",np.round(cv_accuracy,4))
print("## 평균 검증 정확도:",np.mean(cv_accuracy)) 
```
